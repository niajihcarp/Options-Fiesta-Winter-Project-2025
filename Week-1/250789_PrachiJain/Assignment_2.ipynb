```python
# Question 1
arr = []
n = input("Enter the value of n: ")
for i in range(int(n)):
    val = input(f'Enter element no.{(i+1)}: ')
    arr.append(int(val))
print('Max val: ', max(arr))
print('Min val: ', min(arr))
```

```python
# Question 2
str = input('Enter the string: ')
start = 0
end = len(str) - 1
output = 'Palindrome'
while(start<end):
    if str[start] != str[end]:
        output = 'Not a Palindrome'
    start += 1
    end -= 1
print(output)        
```

```python
# Question 4
print("Enter a, b, c in increasing order")
a = input("Enter a: ")
b = input("Enter b: ")
c = input("Enter c: ")
n = input('Enter n: ')
arr = []
for i in range(int(n)):
    val = input(f'Enter element no.{(i+1)}: ')
    arr.append((val))
i=j=0
for curr in range(int(n)):
    if(arr[curr]==a):
        arr[curr] = c
        arr[j] = b
        arr[i] = a
        j +=1
        i+=1
    elif(arr[curr]==b):
        arr[curr] = c
        arr[j] = b
        j+= 1
print(arr)
```
